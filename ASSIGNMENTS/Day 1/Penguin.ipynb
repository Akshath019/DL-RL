```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('penguins_size.csv')

# Step 2: Check the data
print("Rows and columns:", df.shape)
print("Column names:", df.columns.tolist())
print("Missing values:\n", df.isnull().sum())
print("First 5 rows:\n", df.head())


df = df.dropna()


df['body_mass_g'] = df['body_mass_g'].astype(float)
df['flipper_length_mm'] = df['flipper_length_mm'].astype(float)


df['body_mass_kg'] = df['body_mass_g'] / 1000


print("\nAfter cleaning and transformation, rows and columns:", df.shape)
print("First 5 rows:\n", df.head())


avg_body_mass = df.groupby('species')['body_mass_g'].mean()
plt.figure(figsize=(6, 4))
plt.bar(avg_body_mass.index, avg_body_mass.values, color='lightblue')
plt.title('Average Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Average Body Mass (grams)')
plt.show()
plt.close()


plt.figure(figsize=(6, 4))
plt.hist(df['flipper_length_mm'], bins=10, color='lightgreen')
plt.title('Flipper Length Distribution')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Number of Penguins')
plt.show()
plt.close()


island_counts = df['island'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(island_counts, labels=island_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Penguins by Island')
plt.show()
plt.close()
```
